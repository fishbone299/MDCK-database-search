```python
import pandas as pd
import numpy as np
```

```python
comp = pd.read_excel('c')
comp
```

```python
comp["No AB 04 high BA 04"] = np.where(
    comp["Papp AB 04"].isna() & (comp["Papp BA 04"] >= 6),
    "no AB high BA",
    "INV"
)
comp["No AB 02 high BA 02"] = np.where(
    comp["Papp AB 02"].isna() & (comp["Papp BA 02"] >= 6),
    "no AB high BA",
    "INV"
)
comp
```

```python
comp.to_excel("No AB, High BA.xlsx", index=False)
```
